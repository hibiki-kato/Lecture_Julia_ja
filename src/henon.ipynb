```julia
using Plots

function henon_map(x_1, x_2; a = 1.4, b = 0.3)
	x = 1 - a * x_1^2 + b * x_2
	return x
end
```

```julia
n = 1000
trajectory = zeros(n)
trajectory[1] = 0.1
trajectory[2] = 0.1
for i in 1:n-2
	trajectory[i+2] = henon_map(trajectory[i+1], trajectory[i]; a = 1.4, b = 0.0)
end

plot(1:n, trajectory)
```

```julia
scatter(trajectory[10:n-1], trajectory[11:n],
	xlim = (-2, 2),
	ylim = (-2, 2))
```
